```python
import os
import zipfile
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from skimage import filters, morphology, measure
from scipy.signal import savgol_filter
import pandas as pd
import matplotlib as mpl
from google.colab import files # Import for file uploads

# ------------------------------
# STEP 1: UPLOAD FILES
# ------------------------------
print("Please upload your first zip file (e.g., 10cst5cm.zip)")
uploaded = files.upload()
zip10_name = list(uploaded.keys())[0]
print(f"Uploaded: {zip10_name}")

print("\nPlease upload the background image for the first zip (e.g., Img000000.tif)")
uploaded = files.upload()
bg10_name = list(uploaded.keys())[0]
print(f"Uploaded: {bg10_name}")

print("\nPlease upload your second zip file (e.g., 1000cst5cm.zip)")
uploaded = files.upload()
zip1000_name = list(uploaded.keys())[0]
print(f"Uploaded: {zip1000_name}")

print("\nPlease upload the background image for the second zip (e.g., 1000cst 5cm.tif)")
uploaded = files.upload()
bg1000_name = list(uploaded.keys())[0]
print(f"Uploaded: {bg1000_name}")

print("\n--- All files uploaded. Starting analysis... ---")


# ------------------------------
# STEP 2: ANALYSIS PIPELINE
# ------------------------------

# ------------------------------
# GLOBAL SETTINGS
# ------------------------------
pixel_to_mm = 10.0 / 212.0    # calibration from your ruler image
fps = 1800                  # high-speed camera frame rate

# disable LaTeX (important for your environment)
mpl.rcParams.update(mpl.rcParamsDefault)
mpl.rcParams['text.usetex'] = False


# ------------------------------
# FUNCTION: PROCESS ZIP
# ------------------------------
def compute_droplet_diameter(zip_path, bg_path):
    """Returns (time_s, diameter_mm)."""

    extract_dir = zip_path.replace(".zip", "_extract")
    os.makedirs(extract_dir, exist_ok=True)

    print(f"Extracting {zip_path}...")
    with zipfile.ZipFile(zip_path, "r") as z:
        z.extractall(extract_dir)
    print(f"Extraction complete.")

    tiffs = sorted([
        os.path.join(extract_dir, f)
        for f in os.listdir(extract_dir)
        if f.lower().endswith((".tif", ".tiff"))
    ])

    # load background
    bg = np.array(Image.open(bg_path).convert("L")).astype(float)
    h, w = bg.shape

    # detect surface using vertical gradient
    gy = np.abs(np.diff(bg, axis=0))
    surface_row = int(np.argmax(np.mean(gy, axis=1)))

    # crop a band around surface
    top = max(0, surface_row - 60)
    bottom = min(h, surface_row + 5)
    bg_band = bg[top:bottom, :]

    diam_px = []
    print(f"Processing {len(tiffs)} frames...")

    for p in tiffs:
        img = np.array(Image.open(p).convert("L")).astype(float)
        band = img[top:bottom, :]

        diff = bg_band - band
        diff[diff < 0] = 0
        diffn = (diff / diff.max() * 255) if diff.max() > 0 else diff

        try:
            thr = filters.threshold_otsu(diffn.astype(np.uint8))
        except:
            thr = diffn.mean() + 0.5 * diffn.std()

        mask = diffn > thr
        mask = morphology.remove_small_objects(mask, 50)
        mask = morphology.remove_small_holes(mask, 50)

        labels = measure.label(mask)
        props = measure.regionprops(labels.astype(int))

        if len(props) == 0:
            diam_px.append(0)
        else:
            largest = max(props, key=lambda p: p.area)
            _, minc, _, maxc = largest.bbox
            diam_px.append(maxc - minc)

    print("Frame processing complete.")
    # convert px  mm
    diam_mm = np.array(diam_px) * pixel_to_mm

    # convert frame index  time
    time_s = np.arange(len(diam_mm)) / fps

    return time_s, diam_mm


# ------------------------------
# FUNCTION: SAVITZKYGOLAY
# ------------------------------
def smooth_savgol(d_mm, window=21, poly=3):
    return savgol_filter(d_mm, window_length=window, polyorder=poly, mode="interp")


# ------------------------------
# FUNCTION: EXPORT CSV
# ------------------------------
def export_csv(time, raw, smooth, outfile):
    df = pd.DataFrame({
        "time_s": time,
        "diameter_raw_mm": raw,
        "diameter_savgol_mm": smooth
    })
    df.to_csv(outfile, index=False)
    return outfile


# ------------------------------
# FUNCTION: PUBLICATION PLOT
# ------------------------------
def plot_publication(time, curves, labels, title, outfile):
    plt.rc('font', family='serif')
    plt.rc('axes', linewidth=1.5)
    plt.rc('figure', dpi=300)

    plt.figure(figsize=(10,6))

    for d, label in zip(curves, labels):
        plt.plot(time, d, linewidth=2.0, label=label)

    plt.xlabel("Time (s)", fontsize=14, fontweight='bold')
    plt.ylabel("Diameter (mm)", fontsize=14, fontweight='bold')
    plt.title(title, fontsize=17, fontweight='bold')

    plt.grid(True, linewidth=0.4, alpha=0.5)
    plt.legend(fontsize=14)
    plt.tight_layout()

    plt.savefig(outfile, dpi=300)
    print(f"Graph saved to {outfile}")
    plt.show()


# ------------------------------
# MAIN EXECUTION BLOCK
# ------------------------------

# These variables (zip10_name, bg10_name, etc.)
# were set by the upload section at the top of this cell.

print("--- Starting Full Analysis ---")

# --- Process First Dataset ---
print("\nProcessing Dataset 1...")
t10, d10 = compute_droplet_diameter(zip10_name, bg10_name)
sg10 = smooth_savgol(d10)
csv_out1 = "dataset_1_output.csv"
export_csv(t10, d10, sg10, csv_out1)
print(f"Dataset 1 CSV exported to {csv_out1}")

# --- Process Second Dataset ---
print("\nProcessing Dataset 2...")
t1000, d1000 = compute_droplet_diameter(zip1000_name, bg1000_name)
sg1000 = smooth_savgol(d1000)
csv_out2 = "dataset_2_output.csv"
export_csv(t1000, d1000, sg1000, csv_out2)
print(f"Dataset 2 CSV exported to {csv_out2}")

# --- Plot Comparison ---
print("\nGenerating comparison plot...")
plot_name = "droplet_comparison_plot.png"
# Assumes both datasets have the same time axis (t10).
# If they have different lengths, you may need to adjust which time axis is used.
plot_publication(
    t10,
    [sg10, sg1000],
    [zip10_name, zip1000_name], # Labels based on uploaded filenames
    "Droplet Spreading Comparison (SG Smoothed)",
    plot_name
)

print("\n--- Analysis Finished ---")
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

# ------------------------------
# FUNCTION: PUBLICATION PLOT (Corrected)
# ------------------------------
# This new version accepts a list of time arrays ('times')
# instead of just one.
def plot_publication(times, curves, labels, title, outfile):
    # Set plot style
    plt.rc('font', family='serif')
    plt.rc('axes', linewidth=1.5)
    plt.rc('figure', dpi=300)

    plt.figure(figsize=(10,6))

    # Loop over all three lists (times, curves, labels)
    for t, d, label in zip(times, curves, labels):
        plt.plot(t, d, linewidth=2.0, label=label)

    plt.xlabel("Time (s)", fontsize=14, fontweight='bold')
    plt.ylabel("Diameter (mm)", fontsize=14, fontweight='bold')
    plt.title(title, fontsize=17, fontweight='bold')

    plt.grid(True, linewidth=0.4, alpha=0.5)
    plt.legend(fontsize=14)
    plt.tight_layout()

    plt.savefig(outfile, dpi=300)
    print(f"Graph saved to {outfile}")
    plt.show()


# ------------------------------
# MAIN EXECUTION BLOCK (Corrected Call)
# ------------------------------
print("Generating comparison plot with corrected function...")
plot_name = "droplet_comparison_plot.png"

# We are re-using the variables t10, sg10, t1000, and sg1000
# that were calculated in the previous cell.

# The fix is here: We now pass [t10, t1000] as the 'times' argument.
plot_publication(
    [t10, t1000],  # Pass both time arrays
    [sg10, sg1000],
    [zip10_name, zip1000_name], # Labels from uploaded filenames
    "Droplet Spreading Comparison (SG Smoothed)",
    plot_name
)

print("\n--- Plotting Finished ---")
```
