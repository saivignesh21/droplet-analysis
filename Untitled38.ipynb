```python
# ==============================================================
# 🧩 STEP 1: Install & Import
# ==============================================================
!pip install opencv-python-headless scikit-image tqdm matplotlib --quiet

import os, zipfile, cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from skimage import io
from scipy.signal import medfilt
from scipy.interpolate import interp1d
import pandas as pd

# ==============================================================
# 🗂 STEP 2: Upload all zip files (you can upload all at once)
# ==============================================================
from google.colab import files
print("⬆ Upload ALL zip files (10cst_*, 50cst_*, 1000cst_*) together")
uploaded = files.upload()
zip_files = list(uploaded.keys())

# ==============================================================
# ⚙️ Constants
# ==============================================================
FPS = 1800
time_per_frame = 1 / FPS
pixels_per_mm = 36.14
mm_per_pixel = 1 / pixels_per_mm

# ==============================================================
# ⚙️ STEP 3: Droplet Processing Function
# ==============================================================
def ensure_gray(img):
    if len(img.shape) == 3:
        return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img

def process_zip(zip_path):
    """Processes a single zip file and extracts diameter vs time"""
    extract_folder = zip_path.replace('.zip', '')
    os.makedirs(extract_folder, exist_ok=True)
    with zipfile.ZipFile(zip_path, 'r') as zf:
        zf.extractall(extract_folder)

    frames = sorted([f for f in os.listdir(extract_folder) if f.lower().endswith('.tif')])
    if not frames:
        print(f"⚠️ No .tif frames in {zip_path}")
        return None

    diam_mm, times = [], []
    for i, fname in enumerate(tqdm(frames, desc=f"Processing {zip_path}")):
        img = io.imread(os.path.join(extract_folder, fname))
        gray = ensure_gray(img)

        roi = gray[500:800, :]  # adjust as needed
        blur = cv2.GaussianBlur(roi, (5,5), 0)
        _, th = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
        if np.mean(roi[th==255]) > np.mean(roi[th==0]):
            th = cv2.bitwise_not(th)

        kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
        th = cv2.morphologyEx(th, cv2.MORPH_OPEN, kernel)
        th = cv2.morphologyEx(th, cv2.MORPH_CLOSE, kernel)

        cnts, _ = cv2.findContours(th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        if not cnts:
            diam_mm.append(np.nan)
            times.append(i * time_per_frame)
            continue

        c = max(cnts, key=cv2.contourArea)
        eq_d_px = 2 * np.sqrt(cv2.contourArea(c) / np.pi)
        diam_mm.append(eq_d_px * mm_per_pixel)
        times.append(i * time_per_frame)

    diam = np.array(diam_mm)
    idx = np.arange(len(diam))
    if np.any(np.isnan(diam)):
        f = interp1d(idx[~np.isnan(diam)], diam[~np.isnan(diam)], fill_value="extrapolate")
        diam = f(idx)
    diam = medfilt(diam, 5)

    t_ms = np.array(times) * 1000
    D0 = np.median(diam[:5])
    Dmax = np.nanmax(diam)
    Tmax = t_ms[np.argmax(diam)]
    Beta = Dmax / D0

    df = pd.DataFrame({"Time_ms": t_ms, "Diameter_mm": diam})
    csv_name = f"processed_{zip_path.replace('.zip', '.csv')}"
    df.to_csv(csv_name, index=False)

    return {
        "name": zip_path.replace(".zip",""),
        "t_ms": t_ms,
        "diam_mm": diam,
        "D0": D0,
        "Dmax": Dmax,
        "Tmax": Tmax,
        "Beta": Beta
    }

# ==============================================================
# STEP 4: Process All Files
# ==============================================================
results = []
for z in zip_files:
    res = process_zip(z)
    if res: results.append(res)

# ==============================================================
# STEP 5: Time Alignment (resample to common grid)
# ==============================================================
common_time = np.linspace(0, 50, 400)  # 0–50 ms range, 400 points
for r in results:
    f = interp1d(r["t_ms"], r["diam_mm"], fill_value="extrapolate")
    r["interp_diam"] = f(common_time)

# ==============================================================
# STEP 6: Plot Grouped Comparisons
# ==============================================================
def plot_group(viscosity):
    plt.figure(figsize=(8,6))
    for r in results:
        if viscosity in r["name"]:
            plt.plot(common_time, r["interp_diam"], linewidth=2, label=r["name"])
    plt.title(f"{viscosity} – Comparison of Heights")
    plt.xlabel("Time (ms)")
    plt.ylabel("Diameter (mm)")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"{viscosity}_heights_comparison.png")
    plt.show()

for vis in ["10cst", "50cst", "1000cst"]:
    plot_group(vis)

# ==============================================================
# STEP 7: Plot for Constant Height, Different Surfaces
# ==============================================================
heights = ["5cm", "10cm", "20cm", "30cm"]
for h in heights:
    plt.figure(figsize=(8,6))
    for r in results:
        if h in r["name"]:
            plt.plot(common_time, r["interp_diam"], linewidth=2, label=r["name"])
    plt.title(f"Comparison of Surfaces at {h}")
    plt.xlabel("Time (ms)")
    plt.ylabel("Diameter (mm)")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"height_{h}_comparison.png")
    plt.show()
```

```python
# ==============================================================
# 📊 STEP 8: Generate Summary Table for All Data
# ==============================================================

summary_data = []

for r in results:
    # Extract viscosity and height automatically
    name = r["name"].lower()
    viscosity = "10cst" if "10cst" in name else ("50cst" if "50cst" in name else "1000cst")
    height = "5cm" if "5cm" in name else ("10cm" if "10cm" in name else ("20cm" if "20cm" in name else "30cm"))

    summary_data.append({
        "File": r["name"],
        "Viscosity": viscosity,
        "Height": height,
        "D0_mm": round(r["D0"], 3),
        "Dmax_mm": round(r["Dmax"], 3),
        "Beta": round(r["Beta"], 3)
    })

summary_df = pd.DataFrame(summary_data)
summary_df = summary_df.sort_values(["Viscosity", "Height"])
summary_df.reset_index(drop=True, inplace=True)

# Display in Colab and Save
print("\n✅ Droplet Summary Table")
display(summary_df)

summary_df.to_csv("Droplet_Summary_All.csv", index=False)
print("\n📁 Saved as Droplet_Summary_All.csv")
from google.colab import files
files.download("Droplet_Summary_All.csv")

```
